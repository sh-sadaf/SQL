#### Ex. 1 Check the number of rentals for each customer for May 

```python
#defining query
query = "select r.customer_id, count(r.rental_id) as no_of_rentals, c.first_name, c.last_name, date_format(r.rental_date,'%M') as rental_month from sakila.rentals_may as r left join customer as c on c.customer_id = r.customer_id group by customer_id order by count(r.rental_id) desc;"

cursor.execute(query)
```

```python
records = cursor.fetchall()
```

```python
for record in records:
    print(record)
```

```python
# converting data in to pandasdataframe
import pandas as pd
df1 = pd.read_sql_query(query,db)
```

#### Ex. 2 Check the number of rentals for each customer for June

```python
query = "SELECT r.customer_id, COUNT(r.rental_id) AS no_of_rentals,c.first_name,c.last_name, date_format(r.rental_date,'%M') as rental_month FROM sakila.rentals_june AS r LEFT JOIN customer c ON r.customer_id = c.customer_id GROUP BY customer_id ORDER BY COUNT(r.rental_id) DESC;"
```

```python
cursor.execute(query)
```

```python
records = cursor.fetchall()
```

```python
for record in records:
    print(record)
```

```python
df2 =pd.read_sql_query(query,db)
```

#### EX 3-Write a function that checks if customer borrowed more or less films in the month of June as compared to May.

```python
# merging both datas together
df = df1.merge(df2, how = "outer")
```

```python
df.groupby('rental_month')['no_of_rentals'].sum()
```

Customers borrowed more in month of June as compared to May.
